Here "4bf58dd8d48988d1e2931735" is the Art Galley Category ID.

```python
df_venues
venues = json_normalize(result['response']['groups'][0]['items'])
df_venues = venues[['venue.name', 'venue.location.address', 'venue.location.lat', 'venue.location.lng']]
df_venues.columns = ['Name', 'Address', 'Lat', 'Lng']
```

```python


city_lat = np.mean([result['response']['geocode']['geometry']['bounds']['ne']['lat'],
                        result['response']['geocode']['geometry']['bounds']['sw']['lat']])
city_lng = np.mean([result['response']['geocode']['geometry']['bounds']['ne']['lng'],
                        result['response']['geocode']['geometry']['bounds']['sw']['lng']])
maps= folium.Map(location=[city_lat, city_lng], zoom_start=11)
for lat, lng, label in zip(df_venues['Lat'], df_venues['Lng'], df_venues['Name']):
        label = folium.Popup(label, parse_html=True)
        folium.CircleMarker(
            [lat, lng],
            radius=5,
            popup=label,
            color='blue',
            fill=True,
            fill_color='#3186cc',
            fill_opacity=0.7,
            parse_html=False).add_to(maps)  
print("Total number of Art Museums New York = ", result['response']['totalResults'])
print("Showing Top 100")
```
